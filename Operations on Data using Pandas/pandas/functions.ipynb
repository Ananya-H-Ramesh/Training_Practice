```python
import pandas as pd
```

```python
file_name = "https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/amazon_fires.csv"
df = pd.read_csv(file_name,encoding="ISO-8859-1")
```

```python
df
```

```python
new_columns = {
    'ano':'year',
    'estado':'state',
    'mes':'month',
    'numero':'number_of_fires',
    'encontro':'date'
}
df.rename(columns = new_columns,inplace = True)
```

```python
df['number_of_fires']=df['number_of_fires'].str.strip(" Fires").fillna(0).astype(float)
```

```python
df
```

```python
df['month'].unique()
```

Map Functions

```python
month_translations = {
    'Janeiro':'January',
    'Fevereiro':'February',
    'Março':'March',
    'Abril':'April',
    'Maio':'May',
    'Junho':'June',
    'Julho':'July',
    'Agosto':'August',
    'Setembro':'September',
    'Outubro':'October',
    'Novembro':'November',
    'Dezembro':'December'
}

df["month"] = df["month"].map(month_translations)
```

```python
df["month"].unique()
```

```python
df["state"] = df["state"].str.title()
```

```python
import numpy as np

df.pivot_table(values = 'number_of_fires',index = 'state', aggfunc = np.mean)
```
